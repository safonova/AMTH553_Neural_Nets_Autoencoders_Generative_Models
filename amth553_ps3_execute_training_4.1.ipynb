```python
reparam = np.load("reparametrizations.npy")
all_targets = np.load("all_targets.npy")
phate_operator = phate.PHATE()
knn = 10
decay = 20
t = 15
phate_operator.set_params(knn=knn, decay=decay, t=t)
Y_phate = phate_operator.fit_transform(reparam)
```

```python

scprep.plot.scatter2d(Y_phate, figsize=(8, 6), c=all_targets, cmap="Spectral",
                          ticks=False, label_prefix="PHATE", title="PHATE of VAE latent space colored by input digit")
plt.savefig("4_1_PHATE_latent_VAE.png")
```
